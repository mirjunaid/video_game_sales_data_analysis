```python
# Importing the required libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read the data
data = pd.read_csv('vgsales.csv')
data.head()
```

```python
# Clean the data
nan_data = data.isna()
nan_data.head()
```

```python
# Make the columns correct type
data['Year'] = pd.to_numeric(data['Year'])
data['NA_Sales'] = pd.to_numeric(data['NA_Sales'])
data['EU_Sales'] = pd.to_numeric(data['EU_Sales'])
data['JP_Sales'] = pd.to_numeric(data['JP_Sales'])
data['Other_Sales'] = pd.to_numeric(data['Other_Sales'])
data['Global_Sales'] = pd.to_numeric(data['Global_Sales'])
```

# Let's begin with  analysis of the data by drawing graphs for the basic information of the games i.e., Platform, Year published, Genre of the game and publisher..

```python
# Graph for the number of games for each platform
game_platform = data.groupby('Platform')
game_platform_value = game_platform.count()['Name']
game_platform = [pair for pair, df in game_platform]

plt.plot(game_platform, game_platform_value, color='maroon', marker='.', markersize=8)
plt.xticks(game_platform, rotation='vertical', size=10)
plt.grid(True)
plt.title('Graph for the number of games released on each platform', color='red', size=20, weight=750)
plt.xlabel('Platform', color='blue', size=20)
plt.ylabel('Number of games', color='blue', size=16)
plt.show()
```

```python
# Graph for number of games released per year
releasing_year = data.groupby('Year')
games_released = releasing_year.count()['Name']
releasing_year = [pair for pair, df in releasing_year]

plt.plot(releasing_year, games_released, color='green', marker='.', markersize=8)
plt.title('Graph for number of games released per year', color='red', size=20, weight=750)
plt.xlabel('Releasing year', color='blue', size=20)
plt.ylabel('Number of games released', color='blue', size=16)
plt.grid(True)
plt.show()
```

```python
# Graph for number of games released VS Genre of the games
new_colors = ['blue', 'violet', 'orange', 'red', 'purple', 'yellow', 'green', 'maroon', 'gold', 'cyan']
game_genre = data.groupby('Genre')
games_per_genre = game_genre.count()['Name']
game_genre = [pair for pair, df in game_genre]

plt.bar(game_genre, games_per_genre, color=new_colors, edgecolor='black')
plt.xticks(game_genre, rotation='vertical', size=10)

plt.title('Graph for games per Genre', color='red', size=20, weight=750)
plt.xlabel('Game genre', color='blue', size=20)
plt.ylabel('Games per genre', color='blue', size=16)
plt.show()
```

# Let's analyze the sales data for each game

```python
# Graph for sales in America per Genre
na_sales = data.groupby('Genre')
sales_value = na_sales.sum()['NA_Sales']
na_sales = [pair for pair, df in na_sales]

plt.bar(na_sales, sales_value, color=new_colors, edgecolor='black')
plt.xticks(na_sales, rotation='vertical', size=10)
plt.title('Sales in America per Genre', color='red', size=20, weight=750)
plt.xlabel('Genre', color='blue', size=20)
plt.ylabel('Sales in millions', color='blue', size=18)
plt.show()
```

```python
eu_sales = data.groupby('Genre')
sales_value = eu_sales.sum()['EU_Sales']
eu_sales = [pair for pair, df in eu_sales]

plt.bar(eu_sales, sales_value, color=new_colors, edgecolor='black')
plt.xticks(eu_sales, rotation='vertical', size=10)
plt.title('Graph for sales in Europe per Genre', color='red', size=20, weight=750)
plt.xlabel('Genre', color='blue', size=20)
plt.ylabel('Sales in millions', color='blue', size=18)
plt.show()
```

```python
# Graph for sales in japan per Genre
jp_sales = data.groupby('Genre')
sales_value_japan = jp_sales.sum()['JP_Sales']
jp_sales = [pair for pair, df in jp_sales]

plt.bar(jp_sales, sales_value_japan, color=new_colors, edgecolor='black')
plt.xticks(jp_sales, rotation='vertical', size=10)
plt.title('Graph for sales in Japan per Genre', color='red', size=20, weight=750)
plt.xlabel('Genre', color='blue', size=20)
plt.ylabel("Sales in millions", color='blue', size=18)
plt.show()
```

```python
# Graph for sales in other countries per Genre
other_sales = data.groupby('Genre')
sales_value_other = other_sales.sum()['Other_Sales']
other_sales = [pair for pair, df in other_sales]

plt.bar(other_sales, sales_value_other, color=new_colors, edgecolor='black')
plt.xticks(other_sales, rotation='vertical', size=10)
plt.title('Graph for sales per Genre in other countries', color='red', size=20, weight=750)
plt.xlabel('Genre', color='blue', size=20)
plt.ylabel('Sales in millions', color='blue', size=16)
plt.show()
```

```python
global_sales = data.groupby('Genre')
global_sales_value = global_sales.sum()['Global_Sales']
global_sales = [pair for pair, df in global_sales]

plt.bar(global_sales, global_sales_value, color=new_colors, edgecolor='black')
plt.xticks(global_sales, rotation='vertical', size=10)
plt.title('Graph for the Global Sales Per Genre', color='red', size=20, weight=750)
plt.xlabel('Genre', color='blue', size=20)
plt.ylabel('Sales in millions', color='blue', size=18)
plt.show()
```
